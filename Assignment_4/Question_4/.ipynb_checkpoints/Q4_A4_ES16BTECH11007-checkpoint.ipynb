```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Conclusions

4a) PCA tries to map the given data alone the dimensions where there is maximum spread. Among the significant principal components, we selected to most significant principal components and mapped the data along that dimension.

- the plot along those principal components maybe the effective representation of the data, but also the direction which the variance is large.




4b) tSNE and PCA are both methods of dimensionality reduction.
By applying PCA only along two dimensions, we try to best represent the data along those two dimensions, but as can be seen from the plots below, it was able to store the information well, but the spread among the data is not so high to distinguish all the three different classes

- tSNE is a better method than PCA but is computationally costlier. What it does is it tries to minimize the divergence between two distributions, one the distribution that measures pairwise similarities of input objects and the second distribution that measures pairwise similarities of the corresponding low dimesnional points in the embedding, through KL divergence. This way tSNE was better able to visualise the data with more spread in fewer dimensions




4c) For the Same data when tSNE is applied, we get different representation of data each time but in case of PCA we get the same representation for the given data. From the name itself, given " STOCHASTIC " not " DETERMINISTIC ", it must be depending on some initialization or the probability distributions, due to which we get different representation each time.




```python
data = pd.read_csv("iris.data", sep = ",", header = None)
data.head()
```

```python
data.T[0:4].T.head()
```

```python
train_data = data.T[0:4].T
train_data = train_data.values
train_data = np.asarray(train_data)

train_data = np.array(train_data).astype(float)
train_mean = train_data.mean(0)
cov = np.matmul((train_data - train_mean), (train_data - train_mean).T)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(train_data)

pcomponents = pca.fit_transform(train_data)

class1 = []
class2 = []
class3 = []

for i in range(0, len(data)):
    if(data[4][i] == 'Iris-setosa'):
        class1.append(pcomponents[i])
    elif(data[4][i] == 'Iris-versicolor'):
        class2.append(pcomponents[i])
    elif(data[4][i] == 'Iris-virginica'):
        class3.append(pcomponents[i])
    
class1 = np.asarray(class1)
class2 = np.asarray(class2)
class3 = np.asarray(class3)
    
plt.plot(class1.T[0], class1.T[1], 'o')
plt.plot(class2.T[0], class2.T[1], 'o')
plt.plot(class3.T[0], class3.T[1], 'o')
```

```python
# import sklearn
from sklearn.manifold import TSNE
# from sklearn import datasets

X_embedded = TSNE(n_components=2).fit_transform(train_data)
X_embedded.shape

class1 = []
class2 = []
class3 = []

for i in range(0, len(data)):
    if(data[4][i] == 'Iris-setosa'):
        class1.append(X_embedded[i])
    elif(data[4][i] == 'Iris-versicolor'):
        class2.append(X_embedded[i])
    elif(data[4][i] == 'Iris-virginica'):
        class3.append(X_embedded[i])
    
class1 = np.asarray(class1)
class2 = np.asarray(class2)
class3 = np.asarray(class3)
    
plt.plot(class1.T[0], class1.T[1], 'o')
9plt.plot(class2.T[0], class2.T[1], 'o')
plt.plot(class3.T[0], class3.T[1], 'o')
```
