```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Conclusions

```python
data = pd.read_csv("iris.data", sep = ",", header = None)
data.head()
```

```python
data.T[0:4].T.head()
```

```python
train_data = data.T[0:4].T
train_data = train_data.values
train_data = np.asarray(train_data)

train_data = np.array(train_data).astype(float)
train_mean = train_data.mean(0)
cov = np.matmul((train_data - train_mean), (train_data - train_mean).T)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(train_data)

pcomponents = pca.fit_transform(train_data)

class1 = []
class2 = []
class3 = []

for i in range(0, len(data)):
    if(data[4][i] == 'Iris-setosa'):
        class1.append(pcomponents[i])
    elif(data[4][i] == 'Iris-versicolor'):
        class2.append(pcomponents[i])
    elif(data[4][i] == 'Iris-virginica'):
        class3.append(pcomponents[i])
    
class1 = np.asarray(class1)
class2 = np.asarray(class2)
class3 = np.asarray(class3)
    
plt.plot(class1.T[0], class1.T[1], 'o')
plt.plot(class2.T[0], class2.T[1], 'o')
plt.plot(class3.T[0], class3.T[1], 'o')
```

```python
import sklearn
from sklearn.manifold import TSNE
from sklearn import datasets

X_embedded = TSNE(n_components=2).fit_transform(train_data)
X_embedded.shape

class1 = []
class2 = []
class3 = []

for i in range(0, len(data)):
    if(data[4][i] == 'Iris-setosa'):
        class1.append(X_embedded[i])
    elif(data[4][i] == 'Iris-versicolor'):
        class2.append(X_embedded[i])
    elif(data[4][i] == 'Iris-virginica'):
        class3.append(X_embedded[i])
    
class1 = np.asarray(class1)
class2 = np.asarray(class2)
class3 = np.asarray(class3)
    
plt.plot(class1.T[0], class1.T[1], 'o')
plt.plot(class2.T[0], class2.T[1], 'o')
plt.plot(class3.T[0], class3.T[1], 'o')
```
