```python
import numpy as np
import pandas as pd
import json
```

```python
training_data = pd.read_json('train.json')
training_data['all_ingredients'] = training_data['ingredients'].map(".".join)
ingredients_training = training_data.values
```

```python
ingredient_matrix = []
for x in ingredients_training:
    ingredient_matrix.append(list(x[2]))
```

```python
print ingredient_matrix[0]
```

```python
unique_ingredients = set()

for x in ingredient_matrix:
    for z in x:
        unique_ingredients.add(z)
    
    
```

```python
print len(unique_ingredients)
```

```python
ingredient_binary = []

for x in ingredient_matrix:
    ingredient_binary.append( 1*np.isin( list(unique_ingredients), x ))
```

```python
print len(ingredient_binary[0]) 
```

```python
ingredient_binary = 1*ingredient_binary
```
